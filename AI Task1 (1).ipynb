```python
import pandas as pd
```

```python
df=pd.read_csv('data.csv')
```

```python
df.sort_values("Theta", inplace = True)
```

```python
df.drop_duplicates(subset='Theta',keep='first',inplace = True)
```

```python
import math
```

```python
import numpy as np
```

```python
letters=np.array(['A','B','C','D','D0'])
```

```python
for let in letters:
    df[f'{let}-Magnitude of pos. vector (m)']=(df[f'{let}-X (meter)']**2+df[f'{let}-Y (meter)']**2).apply(lambda num:math.sqrt(num))
    df[f'{let}-Magnitude of vel. vector (m/s)']=(df[f'{let}-X Vel (meter/sec)']**2+df[f'{let}-Y Vel (meter/sec)']**2).apply(lambda num:math.sqrt(num))
    df[f'{let}-Magnitude of acc. vector (m/s**2)']=(df[f'{let}-X Acc (meter/sec**2)']**2+df[f'{let}-Y Acc (meter/sec**2)']**2).apply(lambda num:math.sqrt(num))

```
