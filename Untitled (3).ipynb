```python
import pandas as pd
```

```python
df=pd.read_csv('data.csv')
```

```python
df.sort_values("Theta", inplace = True)
```

```python
df
```

```python
df.drop_duplicates(subset='Theta',keep='first',inplace = True)
```

```python
import math
```

```python
df['A-Magnitude of pos. vector (m)']=(df['A-X (meter)']**2+df['A-Y (meter)']**2).apply(lambda num:math.sqrt(num))
```

```python
df['A-Magnitude of vel. vector (m/s)']=(df['A-X Vel (meter/sec)']**2+df['A-Y Vel (meter/sec)']**2).apply(lambda num:math.sqrt(num))
```

```python
df['A-Magnitude of acc. vector (m/s**2)']=(df['A-X Acc (meter/sec**2)']**2+df['A-Y Acc (meter/sec**2)']**2).apply(lambda num:math.sqrt(num))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df['Theta'],df['A-Magnitude of pos. vector (m)'])
plt.ylim(-0.1,0.1)
```
